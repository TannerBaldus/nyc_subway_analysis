#Section 1: Statistcal Analysis

##1.1
Here we will use the Welch's t test with a one tailed p test because we're interested to see if people ride the subway **more** if it's raining. This would make the null hypothesis:

The ridership on days with rain is less than or equal to the mean ridership on days without rain

I would like to 95% confident so the p crtical value will be .05 .

###1.2
The population of that the samples draw from is not normally distributed. However the t test assumes that the sample means are normally distributed, and the central limit therom tells us with large enough sample size the sample means are normally distributed. In my research the I've seen the rule of thumb for large enough sample size to be 30;  which we far surpass with our sample sizes.

##1.3

```python
s, p = sp.stats.ttest_ind(rain,no_rain,equal_var=False)
```

```python
## divde p by 2 to get one tail p value get
p/2
```

```python
(1 - p/2) * 100
```

```python
'Rainy mean {} No Rain mean {}'.format(rain.mean(), no_rain.mean())
```

##1.4
The p value is much smaller than our critical p value, so there is strong evidence to reject the null hyptohesis.
