#Section 1: Statistcal Analysis

##1.1
Here we will use the Welch's two tailed t test to see if there's any diffrence in ridership on rainy days and non non rainy days. The null hypothesis is that *There is no diffrence in means of ridership between rainy days and non rainy days*. Desiring a confidence level of 95% our p-critical value will be .05 .

###1.2
As we can see the population is highly skewed. Which at face value seems to invaidate our use of a t test.

```python
population.hist(bins=50)
```

However since we have a very large number of samples, 42,649 to be exact we know by the central limit therom that the sample distributions are approximatly normally distributed. Therefore the diffrence in means are about normal hence the use of a t test is justifable.

##1.3

```python
s, p = sp.stats.ttest_ind(rain,no_rain,equal_var=False)
```

```python
p
```

```python
(1 - p) * 100
```

```python
'Rainy mean {} No Rain mean {}'.format(rain.mean(), no_rain.mean())
```

##1.4
The p value is much smaller than our critical p value, so there is strong evidence to reject the null hyptohesis.

```python
def add_group_mean_feature(base_df, target_df, categorical_feature,categorical_value, value_feature, new_col_name):
    grouped_series = base_df.groupby(categorical_feature==categorical_value).mean()[value_feature]
    target_df[new_col_name] = grouped_series


group_by_day_weather_df = df.groupby('day_week', as_index=False).mean()
rain_entries = df[df.rain==1].groupby('day_week',as_index=False).mean()[['rain','ENTRIESn_hourly','day_week']]
no_rain_entries = df[df.rain==0].groupby('day_week',as_index=False).mean()[['rain','ENTRIESn_hourly','day_week']]
weekdays = ['Monday','Teusday','Wednesday','Thursday','Friday','Saturday','Sunday']

grouped_entries = pandas.concat([rain_entries, no_rain_entries])
ggplot(aes(x='rain', y='ENTRIESn_hourly',color='day_week'),data=grouped_entries,) + geom_point() + geom_line()+\
scale_x_continuous(breaks=range(0,2),labels=['No Rain','Rain'])+\
  theme_bw()+xlab('')+ylab('Mean Entries')
```

#Section 4: Conclusion

Based on my analysis, more people ride the subway when it **is** raining. I say this because on the results of the two tailed t test and the linear regression. The p value of the two tailed t test was well below the critical value of $.05$ at $4.641402431632967910^{-07}$. Since the p value is so low, we can say something about the direction. Firstly, looking at the diffrence in the means we see that the rainy days have a higher average hourly ridership by about $183$. Also looking at the linear regression model we see that rain accounts for about 24 more hourly riders in the output of the model.

One possible shortcoming of the dataset is that while it has a lot of datapoints it only covers one month, specifcally May 2011. One way I can see this affecting our data in that the weather that the residents are currently acclimated to. For example in in a rainy season you probably won't stay in because of rain, but during the drier season you might decide to stay in on the rainy day.
The summer since you're used to being hot you might think that it's cold around 50 degrees. But after coming out of a harsh winter 50 degrees feels nice. Another shortcoming of our data is that rain is just a binary choice. It would be more useful if instead it was rainfall in inches for in the time since the last recording as a light mist would probably have a diffrent impact than a heavy rain. 

To move on to shortcomings of our analysis, the most obvious one that comes to mind is that the linear regression model assumes that the relationship the variables have are linear. To use temperature as an example again, I would guess that temperature have a parabolic relationship to the output; more people will probably opt to walk with moderate temperatures but opt for the subway when it tempreture tends towards either extreme. This wasn't reflected in our analysis as the coldest average we got was 49 degrees, when it on average get's done to 27 degrees in january in NYC: http://www.currentresults.com/Weather/New-York/Places/new-york-city-temperatures-by-month-average.php
